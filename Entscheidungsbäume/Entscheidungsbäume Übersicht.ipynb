### Artikel Entscheidungsbäume
Erläuterung von grundsätzlichen Begriffen aus dem Artikel

Die **Entropie** ist ein Maß für Unsicherheit der Klassifikation.
Sind die Werte stark einseitig zu einer Klasse verteilt, ist die Entropie klein.

**Klassen** sind die Endergebnisse der Klassifizierung (zb Sport / kein Sport)
**Merkmale** sind Werte nach denen klassifiziert wird(z.B.Temperatur,Regen)

Bei einem **greedy Algorithmus** wird das Lernen in aufeinanderfolgende Teilschritte zerlegt, die einzeln optimiert werden

Der **ID3 Algorithmus** ist der Anfang der Decision trees im ML.
Scy-kit learn arbeitet heutzutage mit dem weiter entwickelten **CART-Algorithmus**

**Random Forests** bestehen aus einer Vielzahl einzelner decision trees mit jeweils zufälligen Merkmalen die demakratisch abstimmen.
-Weniger transparent
-genauer

**Gradient boosted trees** Es werden nacheinander Bäume erstellt die sich immer wieder selbst verbessern

### Handson Jupyter Notebook
-Welche Eigenschaften und Probleme werden adressiert?


Es können nur **orthogonale Grenzen** gezogen werden. Sollen alle Werte der Trainingsdatenmenge richtig klassifiziert müsste es sehr viele Ebenen geben, was dann zwar die Trainignsdaten richtig klassifiziert, insgesamt aber ungenauer wird (overfitting)

Wenn man die gleichen Trainingsdaten mit dem gleichen modell trainiert, können völlig **verschiedene Entscheidungsbäume entstehen**.
Der Grund dafür ist, dass der CART Algorithmus von scikit learn stochastisch vorgeht und nicht den gesamten Datensats auf einmal als Grundlage verwendet.

Decsision trees ohne Restriktionen bezüglich der **minimalen Objekte** pro Ebene können sehr falsche Annahmen treffen. 

### Übungen aus dem Skript

Warum wirken die folgenden Maßnahmen overfitting engegegen?

**Begrenzung der maximalen Tiefe**  
    -Die Bereichsgrenzen bleiben eher grob, Ausreißer werden zwar falsch klassifiziert, aber wenigstens werden alle Daten die sich in den Bereichen befinden richtig der Klasse zugeordnet.    
    -Bei einer zu großen Tiefe lernt das Modell das statistische Rauschen. Man will jedoch eher die grundsätzlichen Regeln trainieren.

**Mindestdatenmenge pro Knoten**  
es wird kein eigener Bereich nur für einen Datenpunkt(Ausreißer generert) Verhindert Trainierung von Rauschen

Beim **pruning** wird der Baum daran gehindert sich voll auszubilden. Das knn vorher (pre-prunging) oder nachher (post-pruning) geschehen

**Ensembles** helfen gegen overfitting, indem das Rauschen dadurch entfernt wird, indem man verschiedene Bäume nimmt die zb alle auf Basis unterschiedlicher Teilmengen trainiert wurden (Random Forests).
